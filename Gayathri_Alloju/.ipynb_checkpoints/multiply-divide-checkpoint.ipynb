```python
from random import randint
import random
import pandas as pd
```

```python
list_of_jsonObjects = []


def replace_parameters(template,parameters):
    for i in parameters:
        k = template.find("$",0,len(template))
        if k!=-1: 
            template = template[:k]+i+template[k+2:]
    return template
```

```python
data = pd.read_excel('input.xlsx', index_col=None, dtype={'Heavy': str,'Light':str},sheet_name='template_13')  
df = pd.DataFrame(data,columns=['Heavy','Light'])
print(df)
template_13 = "Which animal is heavier? $1 or $2 ?"

i=0
while i<100:
    d = {}
            
    # parameters
    list_A = df['Heavy'].tolist()
    list_B = df['Light'].tolist()
    
    obj_A = random.choice(list_A)
    obj_B = random.choice(list_B)
   
        
    parameters = [str(obj_A),str(obj_B)]
    
    #questions
    d["question"] = replace_parameters(template_13,parameters)    
   
    # answer
    ans = obj_A
    option_2 = obj_B

      
    #options
    options_list = [ans, option_2,"both are of equal weight"]
    random.shuffle(options_list)
    
    d["options"] = options_list 
    d["answer"] = ans
    
    list_of_jsonObjects.append(d)

    
    i+=1;
```

```python
data = pd.read_excel('input.xlsx', index_col=None, dtype={'LargeObjects': str,'TinyObjects':str},sheet_name='template_16')  
df = pd.DataFrame(data,columns=['LargeObjects','TinyObjects'])
print(df)
template_16 = "Given a container of 10 kg volume, which object can be filled in more quantity? $1 or $2 ?"

i=0
while i<10:
    d = {}
            
    # parameters
    list_A = df['LargeObjects'].tolist()
    list_B = df['TinyObjects'].tolist()
    
    obj_A = random.choice(list_A)
    obj_B = random.choice(list_B)
   
        
    parameters = [str(obj_A),str(obj_B)]
    
    #questions
    d["question"] = replace_parameters(template_16,parameters)    
   
    # answer
    ans = obj_B
      
    #options
    options_list = [ans, obj_A,"both the objects can be filled in equal quantity "]
    random.shuffle(options_list)
    
    d["options"] = options_list 
    d["answer"] = ans
    
    list_of_jsonObjects.append(d)

    
    i+=1;
```

```python
template_8 = "$1 bought a phone which costed $2 dollars. Life of a phone battery is $3 hours. On using it daily 24/7 , how many days does the battery last?"
i = 5
while i>0:
    d = {}
    price = randint(5000,10000)
    names= random.choice(names_of_employees)
    battery_life = randint(500,1000)
        
    num_of_days = battery_life/24
    
    d["answer"] = int(num_of_days)
 

    battery_life= str(battery_life)
    price= str(price)

    
    parameters = [names,price,battery_life]
    d["question"] = replace_parameters(template_8, parameters)
    d["options"] = [2, int(num_of_days), int(num_of_days)*2, 0]
    
    list_of_jsonObjects.append(d)
    
    i -= 1
```

```python
template_6 = "$1 's father has taken a loan for farming and has to pay $2 every month to repay the loan. He looks after $3 hens of his village for which he will be paid $4 for each per day. Now he earns $5*$6*30 per month. Does he earn enough amount every month that helps him repay the loan?"
i = 5
while i>0:
    d = {}
    to_pay = randint(5000,10000)
    names= random.choice(names_of_employees)
    num_of_hens = randint(10,40)
    payperhen_perday = randint(1,10)
        
    amount = num_of_hens*payperhen_perday*30
    if amount >= to_pay : d["answer"] = "YES" 
    else :  d["answer"] = "NO"  

    to_pay= str(to_pay)
    num_of_hens = str(num_of_hens)
    payperhen_perday = str(payperhen_perday)

    
    parameters = [names,to_pay,num_of_hens,payperhen_perday, num_of_hens, payperhen_perday]
    d["question"] = replace_parameters(template_6, parameters)
    d["options"] = ["YES", "NO"]
    
    list_of_jsonObjects.append(d)
    
    i -= 1
```
